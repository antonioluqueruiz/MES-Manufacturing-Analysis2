```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

def get_operator_by_shift(current_time):
    # Definimos turnos típicos de 8 horas
    hour = current_time.hour
    if 6 <= hour < 14:
        return np.random.choice(["OP-MORNING-01", "OP-MORNING-02"]) # Turno Mañana
    elif 14 <= hour < 22:
        return np.random.choice(["OP-AFTERNOON-01", "OP-AFTERNOON-02"]) # Turno Tarde
    else:
        return np.random.choice(["OP-NIGHT-01", "OP-NIGHT-02"]) # Turno Noche

def run_mes_simulation():
    records = []
    start_time = datetime.now()

    # Parámetros de calidad
    TEMP_LIMIT = 26.0

    for i in range(100):
        timestamp = start_time + timedelta(minutes=i*5)

        # Asignación lógica de operario por turno
        operator = get_operator_by_shift(timestamp)

        # Simulación de sensor
        temp = round(np.random.normal(23, 4), 2)

        # LÓGICA DE MES: Decisión automática basada en datos
        if temp > TEMP_LIMIT:
            action_taken = "ALERTA: Desviación de Temperatura"
            quality_status = "RECHAZADO"
            machine_status = "Stopped"
        else:
            action_taken = "Proceso Normal"
            quality_status = "OK"
            machine_status = "Running"

        records.append({
            "Timestamp": timestamp,
            "Sensor_Temp": temp,
            "Action": action_taken,
            "Quality_Status": quality_status,
            "Machine_State": machine_status,
            "Operator_ID": operator
        })

    df = pd.DataFrame(records)
    df.to_csv("MES_Production_Audit_Trail.csv", index=False, decimal=',')
    print("Simulación de Lógica MES completada. Archivo listo para Power BI.")

run_mes_simulation()
```
